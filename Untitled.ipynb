```python
ignore_list = ['Civil', 'Gut', 'Bend','Arch','Slope','Bar','Crossing', 'Flat', 'Stream', 'Census']


def distance_calc_vectorized(coords_array):
    """Calculate all distances between consecutive points at once"""
    lons, lats = coords_array[:, 0], coords_array[:, 1]
    lat_diffs = np.diff(lats)
    lon_diffs = np.diff(lons)
    avg_lats = (lats[:-1] + lats[1:]) / 2
    lon_diffs_corrected = lon_diffs * np.cos(np.radians(avg_lats))
    return np.sqrt(lat_diffs**2 + lon_diffs_corrected**2) * 111.0
    
def distance_calc(point1, point2):
    """Single point distance calculation (for straight_distance calc)"""
    lon1, lat1 = point1
    lon2, lat2 = point2
    lat_diff = lat2 - lat1
    lon_diff = lon2 - lon1
    lon_diff_corrected = lon_diff * np.cos(np.radians((lat1 + lat2) / 2))
    return np.sqrt(lat_diff**2 + lon_diff_corrected**2) * 111.0
    
def getEntry(trail):    # Loops over the trail object
    
    geometry = trail["geometry"]
    length = trail["lengthmiles"] * 1.60934 # Convert to kilometers
    bicycle_allowed = True
    motorcycle_allowed = True
    if((trail['bicycle'] == None) or (trail['bicycle'] == 'N')):
        bicycle_allowed = False
    if((trail['motorcycle'] == None) or (trail['motorcycle'] == 'N')):
        motorcycle_allowed = False
        
    
    trail_elevations = []
    trail_cumulative_distance = [0]

    prev_coord = None
    total_distance = 0
    ## Check if we deal with a LineString or MultiLine 
    if geometry.geom_type == 'LineString':
        coords = geometry.coords
    elif geometry.geom_type == 'MultiLineString':
        # Multiple lines - concatenate all coordinates
        coords = []
        for line in geometry.geoms:
            coords.extend(list(line.coords))
    else:
        print(f"Unknown geometry type: {geometry.geom_type}")
        return None
    
    ## First get elevation data 
    
    straight_distance = distance_calc(coords[0], coords[-1])
    if straight_distance > 0:
        sinusoty = length / straight_distance 
    else:
        sinusoty = 1.0 
    
 ## VECTORIZED COORDINATE PROCESSING
    coords_array = np.array(list(coords))
    lons, lats = coords_array[:, 0], coords_array[:, 1]

    # Get all elevations at once
    cols, rows = ~transform * (lons, lats)
    trail_elevations = data[rows.astype(int), cols.astype(int)].tolist()

    # Get all distances at once
    if len(coords_array) > 1:
        segment_distances = distance_calc_vectorized(coords_array)
        
        # Build cumulative distance array
        trail_cumulative_distance = [0]
        total_distance = 0
        for distance in segment_distances:
            total_distance += distance
            trail_cumulative_distance.append(total_distance)
    else:
        trail_cumulative_distance = [0]
        total_distance = 0

    print(len(trail_elevations))        
    
    # for long, lat in coords: # Loop over these points to get the elevations in an array
    #         curr_coord = [long, lat]
    #         if(prev_coord is not None): # Add distance from previous point to current
    #             distance_to_prev = distance_calc(curr_coord, prev_coord)
    #             total_distance += distance_to_prev
    #             trail_cumulitive_distance.append(total_distance)
    #         prev_coord = curr_coord            
    #         col, row = ~transform*(long, lat) # Matrix transformation
    #         elevation_val = data[int(row), int(col)]
    #         trail_elevations.append(elevation_val)           
    # print(len(trail_elevations))        
        
    elv_smooth = pd.Series(trail_elevations).rolling(window=3, center=True, min_periods=1).mean().values
    even_distance_m = np.linspace(0, total_distance, len(elv_smooth)) * 1000
    # Now that we have smoothed trail elevations graph, find the slope using this


    slopes = np.gradient(elv_smooth, even_distance_m)*100 # y, x
    slopes = pd.Series(slopes).rolling(window=15, center=True, min_periods=1).mean().values

    max_elevation = max(trail_elevations)
    net_gain = trail_elevations[-1] - trail_elevations[0]
    max_slope = max(slopes) #Steepest part 
    min_elevation = min(trail_elevations)
    average_slope = np.mean(slopes)
    variance = np.var(slopes)
    elevation_changes = len(np.where(np.diff(np.sign(slopes)))[0]) #Counts how many times slope passes zero
    percent_steep = np.mean(slopes > 10)
    
    # Now get the attractions/landmarks
    bounds = geometry.bounds
    margin = 0.1  # degrees - adjust based on how far you want to look
    bbox_mask = (
        (df.prim_long_dec >= bounds[0] - margin) &
        (df.prim_long_dec <= bounds[2] + margin) &
        (df.prim_lat_dec >= bounds[1] - margin) &
        (df.prim_lat_dec <= bounds[3] + margin)
    )
    # Just selects the points that are in this box 
    combined_mask = bbox_mask & (~df['feature_class'].isin(ignore_list))
    filtered_df = df[combined_mask].copy()


# Create GeoDataFrame from filtered data
    landmarks_gdf = gpd.GeoDataFrame(
    filtered_df, 
    geometry=gpd.points_from_xy(filtered_df.prim_long_dec, filtered_df.prim_lat_dec), 
    crs=trails.crs
    )

    # Convert to a projected CRS for accurate distance calculations
    landmarks_gdf_projected = landmarks_gdf.to_crs('EPSG:3857')  # Web Mercator
    geometry_projected = gpd.GeoSeries([geometry], crs=trails.crs).to_crs('EPSG:3857')[0]

# Calculate distances using projected coordinates
    distances = landmarks_gdf_projected.distance(geometry_projected) / 1000
    landmarks_gdf["distance_to_trail"] = distances
    
    water_marks = landmarks_gdf[landmarks_gdf['feature_class'].isin(water_features)]
    population_marks = landmarks_gdf[landmarks_gdf['feature_class'].isin(populated_features)]
    natural_marks = landmarks_gdf[landmarks_gdf['feature_class'].isin(natural_features)]
    
    closest_water = water_marks.nsmallest(4, "distance_to_trail")
    closest_population = population_marks.nsmallest(4, "distance_to_trail")
    closest_natural = natural_marks.nsmallest(4, "distance_to_trail")

    
    trail_info = {
        'trail_name': trail["name"],
        'trail_length (km)': round(length, 2),
        'bikes_allowed':bicycle_allowed,
        'motorcycle_allowed':motorcycle_allowed,
        # Elevation Data
        'elevation_profile':trail_elevations,
        'max_elevation': max_elevation,
        'min_elevation':min_elevation,
        'net_gain': net_gain,
        'average_slope': average_slope,
        'slope_max': max_slope,
        'slope_variance': round(variance, 2),
        'elevation_changes':elevation_changes,
        'percent_steep':round(percent_steep, 3),
        'curvature':round(sinusoty, 2),
        # Landmark and River Data
        'closest_natural': [(row['feature_name'], round(row['distance_to_trail'], 2)) for _, row in closest_natural.iterrows()],
        'closest_cities': [(row['feature_name'], round(row['distance_to_trail'], 2)) for _, row in closest_population.iterrows()],
        'distance_to_water': round(closest_water['distance_to_trail'].iloc[0], 2) if len(closest_water) > 0 else None,
        'distance_to_city': round(closest_population['distance_to_trail'].iloc[0], 2) if len(closest_population) > 0 else None,
        'geometry': geometry
    }
    return trail_info
```

```python
results = []
for i, entry in trails.iterrows():
    if(entry['name'] == None):
        continue;
    result = getEntry(entry)
    results.append(result)
    print(i)
with open("trails.json", "w") as f:
    json.dump(results, f, indent=4, default=str)
```

```python
## KMEANS Clustering for Difficulty Labels

with open("trails.json", "r") as f:
    data = json.load(f)

# Convert to DataFrame
trails = pd.DataFrame(data)
print(trails.columns)
print(len(trails))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.stats import skew


print(trails.columns.tolist())

cluster_map = {
    0:"Easy",
    1:"Moderate",
    2:"Hard"
}
numerical_columns = ['trail_length (km)','max_elevation','min_elevation','net_gain','average_slope','slope_max','slope_variance','elevation_changes','percent_steep','curvature','distance_to_city','distance_to_water']
trails["net_gain"] = trails["net_gain"] / 1000.0   # Convert to km


skewness = trails[numerical_columns].apply(lambda x: skew(x.dropna()))
print("SKEWED: ",skewness)

skewed = ["slope_variance", "curvature", "elevation_changes"]
for col in skewed:
    trails[col] = np.log1p(trails[col])  # log1p handles 0 safely

trails[numerical_columns] = trails[numerical_columns].apply(pd.to_numeric, errors="coerce")
trails = trails.dropna()

X = trails[numerical_columns].values

print(X)

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=3, random_state=42)
trails['cluster'] = kmeans.fit_predict(X_scaled)
trails['difficulty'] = trails['cluster'].map(cluster_map)
trails.drop("difficulty", axis=1)

print(trails['difficulty'].value_counts())
print(trails.groupby("difficulty")[numerical_columns].mean())

trails.to_json("trails_full.json", orient="records", indent=4)


```
